### To generate Question Keyword Answer pairs

From each (subject, domain, predicate, object, range) tuple, we perform Unique Forward relation and Unique Reverse relation to generate QKA pairs.

Unique Forward relation generate QKA pairs in the form of (QK,AK) => ({subject, predicate, range}, object)

Unique Reverse relation generate QKA pairs in the form of (QK,AK) => ({object, predicate, domain}, subject)

The extracted QKA pairs are in question_answer_keywords.txt

```python
QKA_pairs = list()
with open('question_answer_keywords.txt','w+',encoding='utf-8') as wfile:
    with open('sub_pred_obj_triples.txt','r',encoding='utf-8') as file:
        i = 0
        for record in file:
            record = re.split(r'\t+', record)
            if len(record) == 3:
                if record[1] in unique_pred and record[1] in pred_range:
                    pred = record[1][record[1].rfind('.')+1:-1]
                    rng = pred_range[record[1]].rstrip()
                    rng = rng[rng.rfind('.')+1:-1]
                    dom = pred_domain[record[1]].rstrip()
                    dom = dom[dom.rfind('.')+1:-1]
                    record[2] = record[2].rstrip()

                    if record[0] not in entity_names or record[2] not in entity_names:
                        continue
                    i += 1
                    if i%10000 == 0:
                        print(i)
                    QK = (entity_names[record[0]],pred,rng)
                    AK = entity_names[record[2]]
                    QKA_pairs.append((QK,AK))
                    wfile.write('\t'.join(QK)+'\t'+AK+'\n')
                    QK = (entity_names[record[2]],pred,dom)
                    AK = entity_names[record[0]]
                    QKA_pairs.append((QK,AK))
                    wfile.write('\t'.join(QK)+'\t'+AK+'\n')
```
